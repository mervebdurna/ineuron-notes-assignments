```python
n = 5 # to specified row
for i in range(n):
    for j in range(0,i+1):
        print('merve', end = ' ')
    print('\n')

for i in range(n-1,-1,-1):
    for j in range(0,i+1):
        print('merve', end = ' ')
    print('\n')
```

```python
# print the index of 6 and 7
t = (1,2,3,4,5,6,7,8,9)
i = 0
while i < len(t):
    if t[i]==6 or t[i]==7 :
        print(i)
    i=i+1
```

```python
# print the negative index of 6 and 7
t = (6,1,2,3,4,5,6,7,8,9)
i = -1
while i >= -len(t):
    if t[i]==6 or t[i]==7 :
        print(i)
    i=i-1
```
