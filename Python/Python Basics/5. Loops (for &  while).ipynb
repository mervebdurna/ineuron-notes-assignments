```python
#Q6 : try to extract all the numerical data it may be part of dict key or value
num = []
for i in l:
    if type(i) == int:
        num.append(i)
    elif type(i) in [list,set,tuple]:
        for j in i:
            if type(j) == int :
                num.append(j)
    elif type(i) is dict:
        for k,v in i .items():
            if type(k) == int:
                num.append(k)
            if type(v) == int :
                num.append(v)

print(num)
      
```
